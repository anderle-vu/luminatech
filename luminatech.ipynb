# **4. Currency Exchange**

```python
import pandas as pd
import requests

# df3: your cleaned dataframe
# columns used here: 'month' (e.g. '2022-10') and 'currency'

# 1. Make sure 'month' is a proper period (YYYY-MM)
df3['month_dt'] = pd.to_datetime(df3['month'], format='%Y-%m', errors='coerce')
periods = df3['month_dt'].dt.to_period('M').dropna().unique()

# 2. Build FX table: monthly average rates to AUD
currencies = ['USD', 'NZD', 'EUR']
fx_rows = []

for p in periods:
    start_date = p.to_timestamp(how='start').strftime('%Y-%m-%d')
    end_date = p.to_timestamp(how='end').strftime('%Y-%m-%d')
    month_str = p.strftime('%Y-%m')   # e.g. '2022-10'

    for cur in currencies:
        url = f"https://api.frankfurter.app/{start_date}..{end_date}?from={cur}&to=AUD"
        data = requests.get(url).json()
        rates = data.get("rates", {})

        if not rates:
            continue

        aud_values = [v["AUD"] for v in rates.values()]
        avg_rate = sum(aud_values) / len(aud_values)

        fx_rows.append({
            "month": month_str,
            "currency": cur,
            "exchange_rate": avg_rate
        })

fx = pd.DataFrame(fx_rows)

# 3. Merge back to df3 on month + currency
df3 = df3.merge(fx, how='left', on=['month', 'currency'])

# 4. Set AUD = 1
df3.loc[df3['currency'] == 'AUD', 'exchange_rate'] = 1.0

# (optional) drop helper column
df3.drop(columns=['month_dt'], inplace=True)
```

```python
# Calculate conversion
df3['value_sales_aud'] = df3['value_sales'] * df3['exchange_rate']
df3['value_cost_aud'] = df3['value_cost'] * df3['exchange_rate']
df3['profit_aud'] = df3['value_sales_aud'] - df3['value_cost_aud']
```

```python
df3.dtypes
```

### Exchange Rate Processing Note

To ensure all monetary values are normalised to a consistent standard, a monthly average AUD exchange rate was retrieved using the Frankfurter foreign exchange API.  
The `month` column (format: `YYYY-MM`) was used to map each transaction to its corresponding monthly rate.  

- For rows where the currency was `USD`, `NZD`, or `EUR`, the script automatically fetched the average AUD conversion rate for that specific month.  
- For rows where the currency was already `AUD`, the exchange rate was manually set to `1.0` to preserve original values without modification.  

This approach ensures consistent, time-aware currency conversion while avoiding distortions caused by using a single static rate across different periods.

Both `value_sales` and `value_cost` columns contain extreme values (outliers) that could significantly affect the analysis. To ensure data quality and improve the robustness of subsequent analyses, these extreme values were removed.

- For the `value_sales` column, the records with values between -1,000,000 and 1,500,000 were kept.
- For the `value_cost` column, the values less than or equal to 1,000,000 were retained.
- Regarding the `value_quantity` column, although it also contains outliers, these values are more evenly distributed without extreme spikes. Therefore, the entire range of `value_quantity` values were kept without being removed any of its records.

# **6. Skewness**

```python
from scipy.stats import skew
for col in numeric_cols:
    skewness_value = skew(df3[col].dropna())  
    print(f'Skewness of {col}: {skewness_value:.4f}')
```

```python
import numpy as np
from scipy.stats import skew

for col in numeric_cols:
    # Add 1 to avoid log(0), clip negative values to 0
    transformed = np.log1p(df3[col].clip(lower=0))
    
    skewness_value = skew(transformed.dropna())
    print(f'Skewness of log-transformed {col}: {skewness_value:.4f}')
```

```python
print(df3.select_dtypes(include='float').describe())
```

### Skewness 

Skewness was calculated to assess the asymmetry of the numeric variables in the dataset.  
The initial results showed that several variables (such as sales, cost, and quantity) exhibited strong positive skewness, indicating the presence of extreme high values and a long right tail. Such distributions can negatively affect statistical modelling and machine-learning algorithms that assume normality.

To address this, a log transformation (`log1p`) was applied, which reduces the impact of large outliers while preserving the relative structure of the data.  
After transformation, skewness values decreased substantially, resulting in distributions that are closer to symmetrical and therefore more suitable for further analysis.
